```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
img = cv2.imread('doggy.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)
```

```python
t = np.array([[-1,0,-1],
              [0, 6, 0],
              [-1,-1,-1]]) #加起來是一
result = cv2.filter2D(img, -1, t)
result = np.hstack((img, result))
plt.imshow(result)

result= cv2.cvtColor(result, cv2.COLOR_RGB2BGR)
cv2.imwrite('outl.jpg', result)
```
